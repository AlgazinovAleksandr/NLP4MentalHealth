```python
import pickle
import joblib
import pandas as pd
import numpy as np
from catboost import CatBoostClassifier
```

```python
tf_idf_path = 'models/tfidf_vectorizer.pkl'
cb_path = 'models/catboost_model.cbm'
```

```python
with open(tf_idf_path, 'rb') as f:
    tfidf_vectorizer = pickle.load(f)
    print("TF-IDF vectorizer loaded successfully")

catboost_model = CatBoostClassifier()
catboost_model.load_model(cb_path)
print("CatBoost model loaded successfully")
```

```python
def predict_new_text(text, tfidf_vectorizer, catboost_model):
    text_tfidf = tfidf_vectorizer.transform([text])
    prediction = catboost_model.predict(text_tfidf)
    prediction_proba = catboost_model.predict_proba(text_tfidf)
    return prediction[0], prediction_proba[0]
```

```python
pred_label, pred_proba = predict_new_text("One man is feeling very depressed and anxious", tfidf_vectorizer, catboost_model)
pred_label
```
